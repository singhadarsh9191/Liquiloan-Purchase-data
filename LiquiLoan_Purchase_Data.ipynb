```python
#incidence

incidence = df.groupby('Brand')['Incidence'].value_counts()
print(incidence)

fig, ax = plt.subplots()
fig.set_size_inches(10,5)
ax = sns.countplot(df, x='Brand' , hue = 'Incidence')
for p in ax.patches:
    ax.annotate('{:.1f}'.format(p.get_height()), (p.get_x()+0.00, p.get_height()+0.01))
ax.set_xlim(0.7, 5.5)
ax.set_ylim(0,5500)
plt.title("Brand & Incidence")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.legend(title = 'Incidence' , loc = 'upper center', labels = ['Not purchased', 'Purchased'] )
plt.grid()
plt.show()
```

```python
#price revenue.

def price(Data):
  if Data['Brand']==1:
    return Data['Quantity']*Data['Price_1']

  elif Data['Brand']==2:
    return Data['Quantity']*Data['Price_2']

  elif Data['Brand']==3:
    return Data['Quantity']*Data['Price_3']

  elif Data['Brand']==4:
    return Data['Quantity']*Data['Price_4']

  elif Data['Brand']==5:
    return Data['Quantity']*Data['Price_5']

  else:
    return 0


df['update_price']=df.apply(price,axis=1)
```

```python
df.head(10)
```

```python
#compare brand with updated price.

up_pr = df.groupby('Brand')['update_price'].sum()
print(up_pr)

fig, ax = plt.subplots()
up_pr.plot(kind='line',color = 'Red', legend= True)
up_pr.plot(kind='bar',color = 'Blue', legend= True)
ax.set_xlim(0.5,6)
plt.title("Brand & Revenue Comparison")
plt.xlabel("Brand")
plt.ylabel("Total Revenue")

plt.grid()


```

As, we can see here that:

That Maximum revenue is generated by Brand 5 with 29,621 INR, followed by Brand 2 and so on.

```python
#PRmotion with Brand (remaining)

def promotion(Data):
  if Data['Brand']==1 and Data['Promotion_1']==1:
    return 1

  elif Data['Brand']==2 and Data['Promotion_2']==1:
    return 1

  elif Data['Brand']==3 and Data['Promotion_3']==1:
    return 1

  elif Data['Brand']==4 and Data['Promotion_4']==1:
    return 1

  elif Data['Brand']==5 and  Data['Promotion_5']==1:
    return 1

  else:
    return 0


df['updated_promotion']=df.apply(promotion,axis=1)


```
